```python
import pandas as pd
import numpy as np
```

```python
mbta = pd.read_excel(r"C:\\Users\tyler\Documents\mlprjs\mbta.xlsx")
mbta
```

```python
mbta_1 = mbta.drop([1,7,11])
mbta_1
```

```python
mbta_2 = mbta_1.drop(columns=["MBTA Avg Weekday Unlinked Passenger Trips (thousands)"])
mbta_2
```

```python
mbta_3 = mbta_2.melt(id_vars=['Unnamed: 1'], value_vars=mbta_2.columns[0:59],
            value_name="count",var_name = "type")
mbta_3
```

```python
mbta_4 = mbta_3.pivot(columns = 'Unnamed: 1', 
             values = 'count',
             index=['type']).reset_index().\
rename_axis(None, axis=1) 
mbta_4 
```

```python
mbta_4.rename(columns = {'mode':'Date'}, inplace = True)
```

```python
mbta_5 = mbta_4.drop(columns = ["type"])
mbta_5
```

```python
mbta_final = mbta_5.sort_values(by = ["Date"])
mbta_final
```
